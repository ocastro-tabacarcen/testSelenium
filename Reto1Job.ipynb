```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
from gspread.models import Worksheet
sh = gc.open('Data SET')
ws = gc.open('Data SET').sheet1
rows  = ws.get_all_records()
```

```python
wks = sh.add_worksheet(title='resultados',rows='20',cols='20')
```

```python
import pandas as pd
df = pd.DataFrame.from_records(rows)
import numpy as np

df
```

```python
tabla = pd.pivot_table(df,index=('Author','Sentiment'),columns='Country', fill_value='FALSO',aggfunc={'Country':'sum','Theme':'sum'})
tabla
```

```python
wks.update(None,tabla.to_numpy().tolist())
```
